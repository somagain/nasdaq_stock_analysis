```python
meta_rets = Meta['Adj. Close'].pct_change().dropna()
aapl_rets = Apple['Adj. Close'].pct_change().dropna()
amzn_rets = amzn['Adj. Close'].pct_change().dropna()
nflx_rets = netfl['Adj. Close'].pct_change().dropna()
googl_rets = googl['Adj. Close'].pct_change().dropna()
msft_rets = msft['Adj. Close'].pct_change().dropna()



meta_rets.plot(color = 'blue', alpha = 0.2)
amzn_rets.plot(color = 'blue', alpha = 0.2)
aapl_rets.plot(color = 'blue', alpha = 0.2)
nflx_rets.plot(color = 'blue', alpha = 0.2)
googl_rets.plot(color = 'blue', alpha = 0.2)
msft_rets.plot(color = 'blue', alpha = 0.2)
returns.plot(linewidth = 1, color = 'red')
plt.title('TSLA DAILY RETURNS COMPARISON')
```

```python
meta_c_rets = meta_rets.cumsum()
aapl_c_rets = aapl_rets.cumsum()
amzn_c_rets = amzn_rets.cumsum()
nflx_c_rets = nflx_rets.cumsum()
googl_c_rets = googl_rets.cumsum()
msft_c_rets = msft_rets.cumsum()


meta_c_rets.plot(alpha = 0.2, label = 'META', color = 'darkviolet')
amzn_c_rets.plot(alpha = 0.2, label = 'AMZN', color = 'orange')
aapl_c_rets.plot(alpha = 0.2, label = 'AAPL', color = 'black')
nflx_c_rets.plot(alpha = 0.2, label = 'NFLX', color = 'red')
googl_c_rets.plot(alpha = 0.2, label = 'GOOGL', color = 'green')
msft_c_rets.plot(alpha = 0.2, label = 'MSFT', color = 'green')
c_returns.plot(label = 'TSLA', linewidth = 5, color = 'blue')
plt.title('TSLA RETURNS COMPARISON (CUMULATIVE)')
plt.legend()
```

```python
#Market Capitalisation

Meta['MarktCap'] = Meta['Open'] * Meta['Volume']
Apple['MarktCap'] = Apple['Open'] * Apple['Volume']
amzn['MarktCap'] = amzn['Open'] * amzn['Volume']
tsla['MarktCap'] = tsla['Open'] * tsla['Volume']


Meta['MarktCap'].plot(label = 'META')
Apple['MarktCap'].plot(label = 'Apple')
amzn['MarktCap'].plot(label = 'Amazon')
tsla['MarktCap'].plot(label = 'Tesla')

plt.title('Market Cap')
plt.legend()





```

```python
#Percentage Increase in Stock Value
#A percentage increase in stock value is the change in stock comparing that to the previous day. The bigger the value either positive or negative the more volatile the stock is.

#Volatility
Meta['returns'] = (Meta['Close']/Meta['Close'].shift(1)) -1
Apple['returns'] = (Apple['Close']/Apple['Close'].shift(1))-1
amzn['returns'] = (amzn['Close']/amzn['Close'].shift(1)) - 1
tsla['returns'] = (amzn['Close']/tsla['Close'].shift(1)) - 1

Meta['returns'].hist(bins = 100, label = 'META', alpha = 0.5, figsize = (15,7))
Apple['returns'].hist(bins = 100, label = 'Apple', alpha = 0.5)
amzn['returns'].hist(bins = 100, label = 'Amazon', alpha = 0.5)
tsla['returns'].hist(bins = 100, label = 'Tesla', alpha = 0.5)
plt.title('Market Volatility')

plt.legend()
```
